YetGen için bir kahve otomatı yapmak istiyoruz. Bu kahve otomatında aşağıdaki adımları gerçekleştirmesini bekliyoruz. Bu işlemler:

-Kahve Ekle \
-Fiyat Güncelleme \
-Miktarı Değiştirme (ekle/kaldır) \
-Otomatı Açma \
-Otomatı Kapatma \
Bu işlemleri gerçekleştirirken mutlaka class yapısını kullanarak projeyi tamamlamalısınız. Sizi sadece yukarıdaki 5 adım ile sınırlandırmıyoruz bunlara ekstra özellikler ekleyerek projenizi daha da geliştirebilirsiniz.

```python
import time
import random

kahveler = {
    "filtre kahve" : 12,
    "americano" : 18,
    "siyah çay" : 4,
    "latte" : 20,
    "espresso" : 18,
    "cold brew" : 16,
    "chai tea latte" : 25
} 


class Otomat:

  def __init__(self, durum = "açık"):
    self.durum = durum
  

  def otomat_kapat(self):
    print("Otomat kapatılıyor..")
    time.sleep(1.3)
    self.durum = "kapalı"
    

  def onay():
    onay2 = ["h", "e"]
    while True:
      onay = input("Satın alma işlemini onaylıyor musunuz?")
      if onay not in onay2:
        print("Lütfen e veya h giriniz!")
      if onay == "e":
        print("Fiş yazdırılıyor..")
        time.sleep(1.5)
        print("Satın alma işlemi gerçekleşti.")
        break
      elif onay == "h":
        print("Defol git ya beni bu kadar uğraştırdın!!")
        break


  def fis(kahve, boyut, fiyat):
    tarih = time.ctime()
    kahve_isimleri = list(kahveler.keys())
    veriler = f"\n{tarih} tarihinde alındı. \n {kahve_isimleri[kahve].upper()}, {boyut}, {fiyat} TL \n"
    with open("fis", "a") as f:
      f.write(veriler)


  def fiyat2(boyut, kahve):
    kahve_fiyatlari = list(kahveler.values())
    fiyat = kahve_fiyatlari[kahve]
    if boyut == "küçük":
      fiyat -= 2
    elif boyut == "büyük":
      fiyat += 2
    return fiyat


  def menu(self):
    kahve_isimleri = list(kahveler.keys())
    kahve_fiyatlari = list(kahveler.values())

    for i in range(len(kahve_isimleri)):
      print(f"{kahve_isimleri[i].upper()}  : {kahve_fiyatlari[i]} TL/adet")
    print("Fiyatlar orta boyutun fiyatları olup küçük ve büyükte 2 TL değişmektedir. \n")
  

  def satin_alma(self):
    kahve_isimleri = list(kahveler.keys())
    kahve_fiyatlari = list(kahveler.values())
    while True:
      kahve = int(input("Almak istediğiniz içeceğin numarasını giriniz. "))
      if kahve < 0 or kahve > len(kahve_isimleri):
        print("Lütfen geçerli bir sayı giriniz!!")
        continue
      else:
        break
    fiyat = kahve_fiyatlari[kahve]
    boyutlar = ["küçük", "orta", "büyük"]
    while True:
      boyut = input("Hangi boyutta almak istiyorsunuz? ")
      if boyut not in boyutlar:
        print("Lütfen geçerli bir boyut giriniz!!")
        continue
      else:
        break
    fiyat = Otomat.fiyat2(boyut, kahve)
    print(f"Satın almak istediğiniz ürünün fiyatı {fiyat}.")
    Otomat.onay()
    Otomat.fis(kahve, boyut, fiyat)


  def kahve_ekle(self):
    kahve = input("Hangi kahveyi eklemek istiyorsunuz?")
    kahveler[kahve] = random.randint(0,25)


  def fiyat_guncelleme(self):
    while True:
     kahve_isimleri = list(kahveler.keys())
     isim = input("Hangi kahvenin fiyatını güncellemek istiyorsunuz?")
     if isim not in kahve_isimleri:
       print("Kahve listemizde bulunmamaktadır.")
       continue
     else:
        break
    fiyat = int(input(f"Şu an {isim}in fiyatı {kahveler[isim]} TL'dir. Yeni fiyatın ne olmasını istersiniz?"))
    kahveler[isim] = fiyat


  def main(self):
    while True:
      islem = input("Menüyü görmek için 1, kahve satın almak için 2, fiyat güncellemek için 3, kahve eklemek için 4, \notomatı kapatmak için 5 giriniz.")
      if islem == "1":
        self.menu()
      elif islem == "2":
        self.satin_alma()
        break
      elif islem == "3":
        self.fiyat_guncelleme()
        print("Fiyat güncellendi.")
        devam = input("Devam etmek için 1'e, otomatı kapatmak için 0'a basın.")
        if devam == "0":
          self.otomat_kapat()
          break
        elif devam == "1":
          continue
      elif islem == "4":
        self.kahve_ekle()
        print("Kahve eklendi. Yeni kahve listesi: \n")
        self.menu()
        break
      elif islem == "5":
        self.otomat_kapat()
        break
      
otomat = Otomat()
otomat.main()
```
